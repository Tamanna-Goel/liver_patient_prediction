### Liver Patient Prediction

##### Data Overview and Introduction
- Dataset Description: The Liver Disease Prediction dataset is a collection of health records used to predict whether a person has a liver disease or not. The dataset typically contains various medical attributes (such as age, gender, total bilirubin, alkaline phosphatase, etc.) and the target variable indicating whether the individual has a liver disease or not.

- Problem Statement: Predict if a person has liver disease based on medical attributes. This prediction is binary, either the person has the disease (liver disease) or not (no disease).

#### Handling Outliers

```python
plt.figure(figsize=(23,28),facecolor='white')
plotnum=1
for i in data.columns:
    plt.subplot(4,3,plotnum)
    sns.boxplot(x=data[i])
    plotnum+=1
plt.tight_layout()
```

```python
data.loc[data['TotalBilirubin']>15,'TotalBilirubin']=np.median(data['TotalBilirubin'])
```

```python
data.loc[data['DirectBilirubin']>5.0,'DirectBilirubin']=np.median(data['DirectBilirubin'])
```

```python
data.loc[data['AlkalinePhosphotase']>570,'AlkalinePhosphotase']=np.median(data['AlkalinePhosphotase'])
```

```python
data.loc[data['AlamineAminotransferase']>350,'AlamineAminotransferase']=np.median(data['AlamineAminotransferase'])
```

```python
data.loc[data['AspartateAminotransferase']>750,'AspartateAminotransferase']=np.median(data['AspartateAminotransferase'])
```

```python
data.loc[data['TotalProteins']>9,'TotalProteins']=np.median(data['TotalProteins'])
data.loc[data['TotalProteins']<3.5,'TotalProteins']=np.median(data['TotalProteins'])
```

```python
data.loc[data['Ratio']>2.0,'Ratio']=np.median(data['Ratio'])
```

```python
plt.figure(figsize=(23,28),facecolor='white')
plotnum=1
for i in data.columns:
    plt.subplot(4,3,plotnum)
    sns.boxplot(x=data[i])
    plotnum+=1
plt.tight_layout()
```

#### Scaling
- Feature Scaling:- It’s important to scale the features for models that are sensitive to data scaling (e.g., Logistic Regression, SVM).

```python
df1=data.drop('Target',axis=1)
df1
```

```python
# Apply Minmax Scaling
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
df2=sc.fit_transform(df1)
```

```python
df2
```

```python
# converting array into DataFrame
scaled_data=pd.DataFrame(df2,columns=['Age', 'Gender', 'TotalBilirubin', 'DirectBilirubin', 'AlkalinePhosphotase',
       'AlamineAminotransferase', 'AspartateAminotransferase', 'TotalProteins', 'Albumin', 'Ratio'])
```

```python
scaled_data
```

```python
final_df=pd.concat([data.Target,scaled_data],axis=1)
```

### Hyperparameter Tuning
- To improve the performance of these models, you can perform hyperparameter tuning using GridSearchCV or RandomizedSearchCV.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],              # Regularization strength
    'solver': ['liblinear', 'saga', 'newton-cg'],  # Solvers to try
    'max_iter': [100, 200, 300],                # Maximum iterations for convergence
}
```

```python
log_reg = LogisticRegression()
```

```python
grid_search = GridSearchCV(estimator=log_reg, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
```

```python
grid_search.fit(x_train, y_train)
```

```python
print("Best Parameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test)
test_accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy of the best model: {test_accuracy}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Define the parameter grid for Random Forest
param_grid_rf = {
    'n_estimators': [50, 100, 200],  # Number of trees in the forest
    'max_depth': [10, 20, 30],      # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],  # Minimum samples required to split a node
}

# Initialize Random Forest model
rf = RandomForestClassifier()

# Set up GridSearchCV to tune the model
grid_search_rf = GridSearchCV(estimator=rf, param_grid=param_grid_rf, cv=5, scoring='accuracy', n_jobs=-1)

# Fit the model to your data (X_train, y_train)
grid_search_rf.fit(x_train, y_train)

# Output the best parameters and best score
print("Best Parameters for Random Forest: ", grid_search_rf.best_params_)
print("Best Cross-validation Score for Random Forest: ", grid_search_rf.best_score_)

# Evaluate the best model on the test set
best_rf_model = grid_search_rf.best_estimator_
y_pred_rf = best_rf_model.predict(x_test)
print(f"Test Accuracy of the best Random Forest model: {accuracy_score(y_test, y_pred_rf)}")
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

# Initialize GradientBoostingClassifier
gbdt = GradientBoostingClassifier()

# Hyperparameter grid
param_grid_gbdt = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}

# Grid search for hyperparameter tuning
grid_search_gbdt = GridSearchCV(estimator=gbdt, param_grid=param_grid_gbdt, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_gbdt.fit(x_train, y_train)

print(f"Best Parameters: {grid_search_gbdt.best_params_}")
print(f"Test Accuracy of the best model: {test_accuracy}")
```
